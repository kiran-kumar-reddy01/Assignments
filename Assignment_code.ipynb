```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('youth_smoking_drug_data_10000_rows_expanded.csv')
df.head()
```

```python
df.describe()
```

```python
#Smoking prevalence
plt.subplot(1, 2, 1)
plt.hist(df['Smoking_Prevalence'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Smoking Prevalence')
plt.xlabel('Smoking Prevalence')
plt.ylabel('Frequency')

#Drug Experimentation
plt.subplot(1, 2, 2)
plt.hist(df['Drug_Experimentation'], bins=10, color='green',edgecolor = 'black')
plt.title('Distribution of Drug Experimentation')
plt.xlabel('Drug Experimentation')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Line grapgh
df.plot(x='Year', y='Parental_Supervision', kind='line', marker='o', linewidth=1)
plt.xlabel('Year')
plt.ylabel('Parental Supervision')
plt.title('Year-wise Parental Supervision')
plt.ylim(0, 12)
plt.grid()
```
